# **Case Study: Data Preprocessing in Healthcare Analytics**

# Introduction:

In the field of healthcare analytics, hospitals collect vast amounts of patient data. However, real-world datasets often contain missing values, outliers, and features with different scales, making it challenging to apply machine learning models directly. Data preprocessing techniques, such as handling missing values and feature scaling, help in cleaning and standardizing the data for better analysis.

 # Scenario:

A hospital is analyzing patient data to predict the likelihood of diabetes based on several health metrics. However, the dataset contains missing values and features with different ranges. Your task is to preprocess this dataset to make it suitable for further analysis.


# **Steps to Perform**:

```python
import pandas as pd
```

**Perform Feature Scaling**

**1. Using MinMaxScaler (Built-in Library)
Min-Max Scaling transforms data to a range of [0, 1]**

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
df[['Age', 'Glucose_Level']] = scaler.fit_transform(df[['Age', 'Glucose_Level']])
```

(b) Manually Implementing Min-Max Scaling
Min-Max Scaling formula:



image.png

```python
df['Blood_Pressure'] = (df['Blood_Pressure'] - df['Blood_Pressure'].min()) / (df['Blood_Pressure'].max() - df['Blood_Pressure'].min())
```

**2. Standard Scaling (Z-Score Normalization)
Standard Scaling transforms features to have a mean of 0 and a standard deviation of 1.**

(a) Using Built-in StandardScaler

```python
from sklearn.preprocessing import StandardScaler

s_scaler = StandardScaler()
```

```python
df[['BMI']] = s_scaler.fit_transform(df[[ 'BMI']])
```

```python
df['BMI']
```

**(b) Manually Implementing Standard Scaling**
Z-Score formula:

image.png


```python
df['Insulin'] = (df['Insulin'] - df['Insulin'].mean()) / df['Insulin'].std()
```
