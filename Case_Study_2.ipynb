# **Case Study 2: Data Visualization in Sales Analytics**

**Introduction**

In the retail industry, analyzing sales data is crucial for understanding trends, customer behavior, and revenue growth. Data visualization helps businesses gain insights by transforming raw data into meaningful visual representations. In this case study, you will explore and visualize sales data using Matplotlib and Seaborn.

**Scenario:**

A retail company wants to analyze its sales performance over the past year. The dataset contains information about sales transactions, including product categories, sales amounts, customer ratings, and purchase dates. Your task is to generate statistical summaries and visualizations to identify trends and patterns in the sales data.

**Steps to Perform:**

1. Load the dataset using Pandas

```python
import pandas as pd
```

```python
df = pd.read_csv('/content/sales_data.csv')
print("Dataset Head:")
print(df.head())
```

**2. Generate Statistical Summaries**

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Display basic statistics
print("Summary Statistics:")
print(df.describe())
```

**3. Create Data Visualizations**

*a) Histogram of Sales Amounts*

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['Sales_Amount'], bins=10, kde=True, color='blue')
plt.title("Distribution of Sales Amounts")
plt.xlabel("Sales Amount")
plt.ylabel("Frequency")
plt.show()
```

*b) Scatter Plot: Sales Amount vs. Customer Rating*

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df['Customer_Rating'], y=df['Sales_Amount'], hue=df['Product_Category'])
plt.title("Sales Amount vs. Customer Rating")
plt.xlabel("Customer Rating")
plt.ylabel("Sales Amount")
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['number']) # Selects only numeric columns
```

```python
correlation_matrix = numeric_df.corr()
plt.figure(figsize=(6, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
